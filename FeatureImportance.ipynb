```python
with open("featureimportance.txt", "r") as file:
    lines = file.readlines()
    #print(lines)

features_imp = {}
for line in lines:
    items = line.split()
    #print(items)
    key = "_".join(items[0:-1])
    value = items[-1]
    features_imp[key] = value
sorted_feature_imp = dict(sorted(features_imp.items(), key=lambda item: item[1]))
for key in sorted_feature_imp:
    print(key, ",", sorted_feature_imp[key])
```
